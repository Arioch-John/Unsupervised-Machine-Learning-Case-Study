IMPORTING NECESSARY MODULES FOR CLUSTERING, DECOMPOSITION AND PERFORMANCE MEASURE

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python
#                                               --------- GROUP 1 ---------
#       EMPLOYEE MENTAL HEALTH ACROSS DIFFERENT WORKPLACE ENVIRONMENTS

group1_features = dataset[['How many employees does your company or organization have?', 
                                            'Do you currently have a mental health disorder?',
                                            'Do you work remotely?', 
                                            'Benefits Or Help Provided for Mental Health']]
```

```python
# SILHOUETTE SCORE AND WCSS
# 2. Run K-Means and evaluate Silhouette Scores

k_range = range(2, 15)
silhouette_scores =[]
for k in k_range:
    kmeans_eval = KMeans(n_clusters= k, random_state= 42, n_init= 10)
    cluster_labels = kmeans_eval.fit_predict(group1_features)

    # Calculate the average silhouette score for this K
    score = silhouette_score(group1_features, cluster_labels)
    silhouette_scores.append(score)
    print(f"For K = {k}, the average Silhouette Score is: {score:.2f}")
    
#WCSS
 
wcss = []
wcss_range = range (2, 15)

for w in wcss_range:

    kmeans_eval = KMeans(n_clusters=w, random_state=42, n_init= 10)
    elbow= kmeans_eval.fit(group1_features)
    wcss.append(elbow.inertia_)
    print(f"For W = {w}, the WCSS (inertia) is: {elbow.inertia_:2f}")
    
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 5))

# Plot Elbow Method
ax1.plot(wcss_range, wcss, marker="o", color="blue", linestyle="-")
ax1.set_title("Elbow Method (WCSS)")
ax1.set_xlabel("Number of Clusters (K)")
ax1.set_ylabel("WCSS (Inertia)")
ax1.set_xticks(wcss_range)
ax1.grid(True, linestyle="--", alpha=0.5)

# Plot Silhouette Scores
ax2.plot(k_range, silhouette_scores, marker="s", color="green", linestyle="-")
ax2.set_title("Silhouette Scores Graph")
ax2.set_xlabel("Number of Clusters (K)")
ax2.set_ylabel("Average Silhouette Score")
ax2.set_xticks(k_range)
ax2.grid(True, linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Put your feature groups into a dictionary for easy looping
groups = {
    "Group 1: Workplace Environments": group1_features,
    "Group 2: Mental Health Stigma": group2_features,
    "Group 3: Workplace Support": group3_features
}

# Loop through each group to calculate and plot correlation
for group_name, group_data in groups.items():
    
    # 1. Calculate Spearman Correlation Matrix
    # Using .spearman makes it readable and accurate for 0,1,2 ordinal data
    corr_matrix = group_data.corr(method='spearman')
    
    # Print the raw scores in the console for quick reference
    print(f"\n--- Correlation Scores for {group_name} ---")
    print(corr_matrix.round(2))  # Rounded to 2 decimals for readability
    
    # 2. Plot highly readable visual heatmaps
    plt.figure(figsize=(8, 6))
    
    sns.heatmap(
        corr_matrix, 
        annot=True,             # Shows the correlation numbers inside the boxes
        fmt=".2f",              # Limits numbers to 2 decimal places
        cmap="coolwarm",        # Blue = Negative corr, Red = Positive corr
        vmin=-1, vmax=1,        # Fixes the scale from -1 to 1
        linewidths=0.5,         # Adds clear grid lines between blocks
        cbar=True               # Shows the color bar guide
    )
    
    # Adjust titles and layout so long column names do not get cut off
    plt.title(f"Correlation Matrix: {group_name}\n", fontsize=12, fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.impute import SimpleImputer

# COMBINING ALL THE UNIQUE FEATURES 

unique_features = [
    'How many employees does your company or organization have?', 
    'Do you currently have a mental health disorder?',
    'Do you work remotely?', 
    'Benefits Or Help Provided for Mental Health',
    'If you have a mental health issue, do you feel that it interferes with your work when being treated effectively?',
    'Consequences of discussing mental health',
    'History of mental issues',
    'Fear of others behavior, if mentally ill',
    'Saw negative attitude towards mental health problems (Current or Previous Workplace)',
    'Is your anonymity protected if you choose to take advantage of mental health or substance abuse treatment resources provided by your employer?'
]

# EXTRACTING THE FEATURES FROM THE DATASET

joint_features = dataset[unique_features]

# FILLING THE 0s WITH THE MOST FREQUENT VALUES

imputer= SimpleImputer(strategy= 'most_frequent')
imputed_data= imputer.fit_transform(joint_features)


# APPLYING PCA TO VIEW THEM IN LESS DIMENSIONS
pca_joint = PCA(n_components=2, random_state=42)
pca_joint_data = pca_joint.fit_transform(imputed_data)
```

```python
# WCSS AND SILHOUETTE SCORES

wcss = []
silhouette_scores = []
cluster_range = range(2, 15)

# LOOPING FOR EVERY VALUE OF WCSS AND SILHOUETTE SCORE IN THE RANGE

for k in cluster_range:

    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)
    labels = kmeans.fit_predict(pca_joint_data)
    
    # Calculate and print the average silhouette score
    score_joint = silhouette_score(pca_joint_data, labels)
    silhouette_scores.append(score_joint)
    print(f"For K = {k}, the average Silhouette Score is: {score_joint:.2f}")

    
for w in cluster_range:

    kmeans = KMeans(n_clusters=w, random_state=42, n_init= 10)
    elbow_joint= kmeans.fit(pca_joint_data)
    wcss.append(elbow_joint.inertia_)
    print(f"For W = {w}, the WCSS (inertia) is: {elbow_joint.inertia_:2f}")


# PLOTTING

fig, ax = plt.subplots(1, 2, figsize=(16, 5))
sns.set_theme(style="whitegrid")

# WCSS

sns.lineplot(x=list(cluster_range), y=wcss, marker='o', color='#1f77b4', ax=ax[0], linewidth=2.5)
ax[0].set_title('The Elbow Method (WCSS)', fontsize=14, pad=10, weight='bold')
ax[0].set_xlabel('Number of Clusters (K)', fontsize=12)
ax[0].set_ylabel('Within-Cluster Sum of Squares (WCSS)', fontsize=12)
ax[0].set_xticks(list(cluster_range))

# SILHOUETTE SCORES

sns.lineplot(x=list(cluster_range), y=silhouette_scores, marker='o', color='#ff7f0e', ax=ax[1], linewidth=2.5)
ax[1].set_title('Silhouette Score Analysis', fontsize=14, pad=10, weight='bold')
ax[1].set_xlabel('Number of Clusters (K)', fontsize=12)
ax[1].set_ylabel('Silhouette Score', fontsize=12)
ax[1].set_xticks(list(cluster_range))

plt.tight_layout()
plt.show()
```

```python
from sklearn.impute import SimpleImputer

# COMBINING ALL THE UNIQUE FEATURES 

unique_features = [
    'How many employees does your company or organization have?', 
    'Do you currently have a mental health disorder?',
    'Do you work remotely?', 
    'Benefits Or Help Provided for Mental Health',
    'If you have a mental health issue, do you feel that it interferes with your work when being treated effectively?',
    'Consequences of discussing mental health',
    'History of mental issues',
    'Fear of others behavior, if mentally ill',
    'Saw negative attitude towards mental health problems (Current or Previous Workplace)',
    'Is your anonymity protected if you choose to take advantage of mental health or substance abuse treatment resources provided by your employer?'
]

# EXTRACTING THE FEATURES FROM THE DATASET

joint_features = dataset[unique_features]

# FILLING THE 0s WITH THE MOST FREQUENT VALUES

imputer= SimpleImputer(strategy= 'most_frequent')
imputed_data= imputer.fit_transform(joint_features)

# APPLYING PCA

pca_joint = PCA(n_components=2, random_state=42)
pca_joint_data = pca_joint.fit_transform(imputed_data)

# CLUSTERING

kmeans_master = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init=10)
dataset['Overall Clusters'] = kmeans_master.fit_predict(pca_joint_data)

# PLOTTING

plt.figure(figsize= (8, 5))
plt.scatter(
    pca_joint_data[:, 0],
    pca_joint_data[:, 1],
    c= dataset['Overall Clusters'],
    cmap= 'viridis',
    alpha= 0.8,
    s= 30
)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Joint Features Clusters Visualised in PCA Space')
plt.colorbar(label= 'Cluster')
plt.show()

# INFORMATION RETENTION

variance_captured = sum(pca_joint.explained_variance_ratio_) * 100
print(f"Total information retained from the 10 columns: {variance_captured:.2f}%")

# VERIFICATION SCORE

score = silhouette_score(pca_joint_data, dataset['Overall Clusters'])
print(f"Joint Model Silhouette Score: {score:.3f}\n")
```
